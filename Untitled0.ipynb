```python
# ==========================================
# MINI PROJECT: E-COMMERCE STATISTICAL ANALYSIS
# ==========================================

# Import numpy for numerical calculations
import numpy as np

# Import pandas for data creation and manipulation
import pandas as pd

# Import matplotlib for plotting graphs
import matplotlib.pyplot as plt

# Import seaborn for advanced statistical plots
import seaborn as sns

# Import statistical functions from scipy
from scipy import stats

# Import Linear Regression model
from sklearn.linear_model import LinearRegression

# ------------------------------------------
# STEP 1: SET RANDOM SEED
# ------------------------------------------

# Setting seed ensures same random values every time
np.random.seed(42)

# ------------------------------------------
# STEP 2: GENERATE DATASET
# ------------------------------------------

# Total number of orders
n = 1000

# Create DataFrame with realistic e-commerce data
data = pd.DataFrame({

    # Unique order ID (Nominal data)
    "order_id": range(1, n + 1),

    # Warehouse locations (Nominal scale)
    "warehouse": np.random.choice(
        ["North", "South", "East", "West"], n),

    # Product categories (Nominal scale)
    "product_cat": np.random.choice(
        ["Electronics", "Fashion", "Grocery"], n),

    # Payment type (Nominal scale)
    "payment_type": np.random.choice(
        ["COD", "Prepaid"], n, p=[0.4, 0.6]),

    # Delivery days (Ratio scale)
    "delivery_days": np.random.normal(
        5, 1.5, n).clip(1),

    # Order value in dollars (Ratio scale)
    "order_value": np.random.normal(
        250, 80, n).clip(50),

    # Discount percentage (Ratio scale)
    "discount_percent": np.random.randint(0, 40, n),

    # Customer rating (Ordinal scale)
    "rating": np.random.randint(1, 6, n),

    # Order returned or not (Nominal scale)
    "returned": np.random.choice(
        ["Yes", "No"], n, p=[0.15, 0.85])
})

# ------------------------------------------
# STEP 3: DATA VISUALIZATION
# ------------------------------------------

# Bar Chart: Orders by warehouse
data["warehouse"].value_counts().plot(kind="bar")
plt.title("Orders by Warehouse")
plt.xlabel("Warehouse")
plt.ylabel("Number of Orders")
plt.show()

# Histogram: Delivery days
plt.hist(data["delivery_days"], bins=20)
plt.title("Histogram of Delivery Days")
plt.xlabel("Delivery Days")
plt.ylabel("Frequency")
plt.show()

# Pie Chart: Payment type distribution
data["payment_type"].value_counts().plot(
    kind="pie", autopct="%1.1f%%")
plt.title("Payment Type Distribution")
plt.ylabel("")
plt.show()

# Line Graph: Average delivery days vs discount
avg_delivery = data.groupby(
    "discount_percent")["delivery_days"].mean()

plt.plot(avg_delivery.index, avg_delivery.values, marker="o")
plt.title("Average Delivery Days vs Discount")
plt.xlabel("Discount %")
plt.ylabel("Average Delivery Days")
plt.show()

# Boxplot: Order value by product category
sns.boxplot(x="product_cat", y="order_value", data=data)
plt.title("Order Value by Product Category")
plt.show()

# Frequency polygon: Delivery days
freq, bins = np.histogram(data["delivery_days"], bins=10)
midpoints = (bins[:-1] + bins[1:]) / 2

plt.plot(midpoints, freq, marker="o")
plt.title("Frequency Polygon of Delivery Days")
plt.show()

# Ogive: Cumulative delivery days
cum_freq = np.cumsum(freq)

plt.plot(midpoints, cum_freq, marker="o")
plt.title("Ogive (Cumulative Delivery Days)")
plt.show()

# ------------------------------------------
# STEP 4: MEASURES OF CENTRAL TENDENCY
# ------------------------------------------

mean_dd = data["delivery_days"].mean()        # Mean
median_dd = data["delivery_days"].median()    # Median
mode_dd = data["delivery_days"].mode()[0]     # Mode

# ------------------------------------------
# STEP 5: WEIGHTED MEAN
# ------------------------------------------

weighted_mean = np.average(
    data["order_value"],
    weights=data["discount_percent"])

# ------------------------------------------
# STEP 6: COMBINED MEAN (NORTH + SOUTH)
# ------------------------------------------

combined_mean = data[
    data["warehouse"].isin(["North", "South"])
]["delivery_days"].mean()

# ------------------------------------------
# STEP 7: MEASURES OF DISPERSION
# ------------------------------------------

range_dd = data["delivery_days"].max() - data["delivery_days"].min()
variance_dd = data["delivery_days"].var()
std_dd = data["delivery_days"].std()

Q1 = data["delivery_days"].quantile(0.25)
Q3 = data["delivery_days"].quantile(0.75)
IQR = Q3 - Q1

cv = std_dd / mean_dd   # Coefficient of Variation

# ------------------------------------------
# STEP 8: MOST CONSISTENT WAREHOUSE
# ------------------------------------------

consistency = data.groupby(
    "warehouse")["delivery_days"].std()

most_consistent_warehouse = consistency.idxmin()

# ------------------------------------------
# STEP 9: SKEWNESS & KURTOSIS
# ------------------------------------------

skewness = stats.skew(data["order_value"])
kurtosis = stats.kurtosis(data["order_value"])

sns.histplot(data["order_value"], kde=True)
plt.title("Order Value Distribution with KDE")
plt.show()

# ------------------------------------------
# STEP 10: CORRELATION ANALYSIS
# ------------------------------------------

pearson_corr = data["discount_percent"].corr(
    data["order_value"], method="pearson")

spearman_corr = data["delivery_days"].corr(
    data["rating"], method="spearman")

plt.scatter(data["discount_percent"], data["order_value"])
plt.title("Discount % vs Order Value")
plt.xlabel("Discount %")
plt.ylabel("Order Value")
plt.show()

# ------------------------------------------
# STEP 11: LINEAR REGRESSION
# ------------------------------------------

X = data[["discount_percent"]]      # Independent variable
Y = data["order_value"]             # Dependent variable

model = LinearRegression()
model.fit(X, Y)

intercept = model.intercept_
slope = model.coef_[0]

predicted_value_20 = model.predict([[20]])[0]

# ------------------------------------------
# STEP 12: PROBABILITY
# ------------------------------------------

P_returned = (data["returned"] == "Yes").mean()

P_prepaid_not_returned = (
    (data["payment_type"] == "Prepaid") &
    (data["returned"] == "No")
).mean()

# ------------------------------------------
# STEP 13: BINOMIAL DISTRIBUTION
# ------------------------------------------

binomial_prob = stats.binom.pmf(3, 8, P_returned)

# ------------------------------------------
# STEP 14: POISSON DISTRIBUTION
# ------------------------------------------

poisson_prob = stats.poisson.pmf(6, 4)

# ------------------------------------------
# STEP 15: NORMAL DISTRIBUTION
# ------------------------------------------

prob_delivery_gt_7 = 1 - stats.norm.cdf(
    7, mean_dd, std_dd)

# ------------------------------------------
# STEP 16: SAMPLING
# ------------------------------------------

# Simple random sample
simple_random_sample = data.sample(100)

# Stratified sample by warehouse
stratified_sample = data.groupby(
    "warehouse").sample(25)

# ------------------------------------------
# STEP 17: CENTRAL LIMIT THEOREM
# ------------------------------------------

sample_means = [
    data["order_value"].sample(50).mean()
    for _ in range(500)
]

plt.hist(sample_means, bins=30)
plt.title("Central Limit Theorem Demonstration")
plt.show()

# ------------------------------------------
# STEP 18: STANDARD ERROR
# ------------------------------------------

standard_error = std_dd / np.sqrt(100)
```
